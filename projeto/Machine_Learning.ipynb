### Case: Score de Cr√©dito dos Clientes

Voc√™ foi contratado por um banco para conseguir definir o score de cr√©dito dos clientes. Voc√™ precisa analisar todos os clientes do banco e, com base nessa an√°lise, criar um modelo que consiga ler as informa√ß√µes do cliente e dizer automaticamente o score de cr√©dito dele: Ruim, Ok, Bom

```python
# An√°lisando 1 cliente:
# Se o cliente √©: Bom, Ok ou Ruim (nota de cr√©dito dele/score)

# Bibliotecas üìö:
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import (RandomForestClassifier)
from sklearn.neighbors import (KNeighborsClassifier)
from sklearn.metrics import accuracy_score

# Passo a Passo abaixo:
```

```python
# 1¬∫ Passo - Importar a base de dados;
base_dados = pd.read_csv("clientes.csv") # base antiga
base_dados2 = pd.read_csv("novos_clientes.csv") #base nova
display(base_dados)
display(base_dados2)
display(base_dados.info())
base_dados_update1 = base_dados.drop(columns="id_cliente")
display(base_dados_update1)

# int ==> n√∫meros inteiros;
# float ==> n√∫meros flutuantes;
# object ==> strings/textos.
```

```python
# 2¬∫ Passo - Tratar e preparar os dados pra IA;

codificador_profissao = LabelEncoder()
base_dados_update1["profissao"] = codificador_profissao.fit_transform(base_dados_update1["profissao"])

codificador_profissao2 = LabelEncoder()
base_dados2["profissao"] = codificador_profissao2.fit_transform(base_dados2["profissao"])

codificador_mix_credito = LabelEncoder()
base_dados_update1["mix_credito"] = codificador_mix_credito.fit_transform(base_dados_update1["mix_credito"])

codificador_mix_credito2 = LabelEncoder()
base_dados2["mix_credito"] = codificador_mix_credito2.fit_transform(base_dados2["mix_credito"])
codificador_comportamento_pagamento = LabelEncoder()
base_dados_update1["comportamento_pagamento"] = codificador_comportamento_pagamento.fit_transform(base_dados_update1["comportamento_pagamento"])

codificador_comportamento_pagamento2 = LabelEncoder() 
base_dados2["comportamento_pagamento"] = codificador_comportamento_pagamento2.fit(base_dados2["comportamento_pagamento"])


x = base_dados_update1.drop(columns="score_credito")
y = base_dados_update1["score_credito"]
x_treino, x_teste, y_treino, y_teste = train_test_split(x,y,test_size=0.3)


#colunas:
# Profiss√£o (cientista, marceneiro, predreiro, pintor...);
    #substitua as profissoes por n√∫meros (1,2,3,4,5,6,7... enfim)
# mix_cr√©dito ;
# comportamento_pagamento;
# score de cr√©dito (PODE SER TEXTO‚úÖ)

# quero prever a coluna 'score' de cr√©ditos
# x = quem eu usarei p fzr a previsao (base_dados_update1.drop(columns="score_credito"))
# y = quem eu preverei (base_dados_update1["score_credito"])
```

```python
# 3¬∫ Passo - Criar o modelo de IA ==> Ruim, Ok e Boa;
#a) √Årvore de decisao (RandomForest)
#b) Nearest Neighbors (Vizinhos Pr√≥ximos)

# 1 - Importar a 'IA';

# 2 - Criar a 'IA'(modelo);
modelo_arvore_decisao = RandomForestClassifier()
modelo_knn = KNeighborsClassifier()
# 3 - Treinar a 'IA'(modelo).
modelo_arvore_decisao.fit(x_treino,y_treino)
modelo_knn.fit(x_treino,y_treino)
```

```python
# 4¬∫ Passo - Avaliar e escolher o melhor modelo;
previsao_arvore = modelo_arvore_decisao.predict(x_teste)
previsao_knn = modelo_arvore_decisao.predict(x_teste)

display(f"{accuracy_score(y_teste, previsao_arvore)*100:.2f}%")
display(f"{accuracy_score(y_teste, previsao_knn)*100:.2f}%")
```

```python
# 5¬∫ Passo - Fazer novas previs√µes.
previsao_arvore_novo = modelo_arvore_decisao.predict(x_teste)
previsao_knn_novo = modelo_knn.predict(x_teste)


display(f"{accuracy_score(y_teste, previsao_arvore_novo)*100:.2f}%")
display(f"{accuracy_score(y_teste, previsao_knn_novo)*100:.2f}%")

display(base_dados2)
```
